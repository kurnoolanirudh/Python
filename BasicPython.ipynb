```python
def sieve_of_eratosthenes(max_num):
    bool_lst = [True] * max_num
    bool_lst[0], bool_lst[1] = False, False
    for i in range(max_num):
        if bool_lst[i] == True:
            j = 2 * i
            while j < max_num:
                bool_lst[j] = False 
                j += i

    primes = []
    for i in range(max_num):
        if bool_lst[i] == True:
            primes.append(i)

    return primes
```

4 main pillars of OOPS

(1) Encapsulation = encapsulating the data and methods of a object into a single unit

```python
class Person:
    def __init__(self, first_name, last_name, age, gender):
        # __init__() is the constructor 
        # this function is called whenever a object of the class is created
        self.first_name = first_name 
        self.last_name = last_name
        self.age = age 
        self.gender = gender

    # a custom string representation of this object 
    # the below function is called when the object is printed 
    def __str__(self):
        return f"Person[name = {self.full_name()} ; age = {self.age} ; gender = {self.gender}]"

    def __repr__(self):
        return f"Person(name = {self.full_name()} ; age = {self.age} ; gender = {self.gender})"

    def full_name(self):
        return f"{self.first_name} {self.last_name}"
```

(2) Inheritance = A child class can inherit the methods, attributes of the parent class

```python
from typing import override

class Student(Person):
    def __init__(self, first_name, last_name, age, gender, grade, school_name):
        super().__init__(first_name, last_name, age, gender)
        self.grade = grade
        self.school_name = school_name 

    # method oveloading (runtime polymorphism)
    def __repr__(self):
        return f"Student(name = {self.full_name()} ; age = {self.age} ; gender = {self.gender} ; grade = {self.grade} ; school = {self.school_name })"

    @override # a type annotation that is used to annotate the fact that the below method is supposed to override a method...
    #...with the same name in the parent class
    def __str__(self):
        return f"Student[name = {self.full_name()} ; age = {self.age} ; gender = {self.gender} ; grade = {self.grade} ; school = {self.school_name }]"
```

abstraction = hiding complex implementation details 

```python
from abc import abstractmethod

class Shape:
    def __init__(self, name):
        self.name = name 

    @abstractmethod
    def area(self):
        NotImplemented

    @abstractmethod
    def perimeter(self):
        NotImplemented


class Rectangle(Shape):
    def __init__(self, name, length, breadth):
        super().__init__(name)
        self.length = length 
        self.breadth = breadth 

    def area(self):
        return self.length * self.breadth 

    def perimeter(self):
        return 2 * (self.length + self.breadth)

class Square(Shape):
    def __init__(self, name, side_length):
        super().__init__(name)
        self.side_length = side_length

    def area(self):
        return self.side_length ** 2 

    def perimeter(self):
        return 4 * self.side_length
```

```python
s = Shape("empty")
r = Rectangle("rectangle", 3.5, 4.5)
sq = Square("square", 2.7)
```

```python
def print_shape(s : Shape):
    print(f"name = {s.name} ; area = {s.area()} ; perimeter = {s.perimeter()}")
```

```python
print_shape(s)
```

```python
print_shape(r)
```

```python
print_shape(sq)
```

Python has operator overloading

```python
class vector:
    def __init__(self, x, y, z):
        self.x = x 
        self.y = y 
        self.z = z 

    def __repr__(self):
        return f"vector[{self.x}, {self.y}, {self.z}]"

    def __str__(self):
        return f"vector({self.x}, {self.y}, {self.z})"

    def __add__(self, other):
        return vector(self.x + other.x, self.y + other.y, self.z + other.z)

    def __sub__(self, other):
        return vector(self.x - other.x , self.y - other.y, self.z - other.z)

    def __mul__(self, other):
        return vector(self.x * other.x, self.y * other.y, self.z * other.z)

    def __truediv__(self, other):
        return vector(self.x / other.x, self.y / other.y, self.z / other.z)

    def __floordiv__(self, other):
        return vector(self.x // other.x , self.y // other.y, self.z // other.z)
```

getter and setters with @property , @attribute.setter

```python
class Fruit:
    def __init__(self, name):
        self.__name = name 

    def __repr__(self):
        return f"fruit name = {self.__name}"

    def __str__(self):
        return f"fruit name = {self.__name}"

    @property
    def fruit_name(self):
        print("fruit name was accessed")
        return self.__name 
    
    @fruit_name.setter 
    def fruit_name(self, new_name):
        print(f"fruit name is updated from {self.__name} to {new_name}")
        self.__name = new_name

    @fruit_name.deleter 
    def fruit_name(self):
        print(f"fruit name deleted")
        self.__name = None
```

```python
f = Fruit("banana")
```

thanks to the property decorator the fruit_name function can be accessed as its a property

```python
f.fruit_name
```

@staticmethod, = nothing object or class related stuff is automatically passed in 

```python
class Math:
    @staticmethod
    def newton_sqrt(num : int) -> float:
        x = float(num)
        n = x 
        i = 0
        while (x*x - n > 0) and i < 100:
            x = (x + (n / x)) / 2
            i += 1
        return x 
    
    @staticmethod
    def sum(*args):
        res = 0
        for i in args:
            res += i 
        return res
    
    @staticmethod
    def mul(*args):
        res = 1 
        for i in args:
            res *= i 
        return res
```

```python
Math.newton_sqrt(2)
```

```python
Math.sum(1, 2, 3, 4, 5)
```

```python
Math.mul(1, 2, 3, 4, 5)
```

@classmethod = class itself is passed by default, class variables = a variable that is part of the circle i.e. part of every object

```python
class Circle:
    pi = 3.14 

    def __init__(self, radius):
        self.radius = radius 

    def area(self):
        return Circle.pi * self.radius * self.radius
    
    def perimeter(self):
        return 2 * Circle.pi * self.radius
    
    @classmethod
    def create_cirlce(cls, radius):
        return cls(radius)
```

```python
c1 = Circle(3.0)
```

```python
print(c1.area())
print(c1.perimeter())
```

```python
c2 = Circle.create_cirlce(4.0)
```

```python
print(c2.area())
print(c2.perimeter())
```

Generator Inside A Generator

```python
from typing import Any


def wrapper(g : Generator[str, None, None], max_num) -> Generator[Any, None, None]:
    yield "start"
    for i in g:
        yield i
    yield "end"
```

```python
max_num = 4
w = wrapper(number_gen(max_num), max_num)
for i in range(max_num):
    print(next(w))
```

the above thing can be done in a easier way using yield from keyword

```python
from typing import Any

def wrapper(g : Generator[str, None, None], max_num) -> Generator[Any, None, None]:
    yield "start"
    yield from g
    yield "end"

max_num = 4
w = wrapper(number_gen(max_num), max_num)
for i in range(max_num):
    print(next(w))
```

custom decorators

```python
from typing import Callable
from functools import wraps 
from time import perf_counter, sleep

def get_time(func : Callable) -> Callable:
    @wraps(func) # if a function(foo) has this custom decorator. the wraps decorator ensures that when foo.__name__ is accessed foo is returned and not wrapper 
    def wrapper(*args, **kwargs) -> None:
        start_time : float = perf_counter()
        func(*args, **kwargs)
        end_time : float = perf_counter()
        print(f"time taken by {func.__name__} is {end_time - start_time:.3f}s")
    return wrapper
```

```python
@get_time
def sum(a : int, b : int) -> int:
    sleep(2.5)
    print(f"{a} + {b} = {a + b}")
```

lrucache and cache decorators 

Memoization is an optimization technique that stores the results of expensive function calls and returns the cached result when the same inputs occur again. Both decorators implement memoization to speed up function execution.

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize = 128, typed=True) # max_size defines the max number of entries that can be cached(it should preferably be a power of 2)
def factorial(n):
    if n == 0:
        return 1 
    else:
        return n * factorial(n - 1)
```

```python
factorial(5)
```

```python
factorial(10)
```

```python
factorial.cache_info()
```

```python
factorial.cache_parameters()
```

```python
factorial.cache_clear()
```

```python
from functools import cache 
```

```python
@cache
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)
```

```python
fact(5)
```

```python
fact(10)
```

```python
fact.cache_info()
```

```python
fact.cache_parameters()
```

```python
fact.cache_clear()
```
