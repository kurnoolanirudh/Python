bytes(immutable collection of bytes)

```python
bt = b"Hello, World"
```

```python
type(bt)
```

To decode the above bytes type to a string

```python
s = bt.decode()
print(s)
print(type(s))
```

To encode a string to bytes type

```python
bt = s.encode("utf-8")
```

```python
print(bt)
print(type(bt))
```

bytearray = like bytes but mutable

```python
bt_ar = bytearray("Hello, World", "utf-8")
```

```python
print(bt_ar)
print(type(bt_ar))
```

```python
bt_ar[0] += 1
```

```python
bt_ar
```

To decode a byte array to a string

```python
s = bt_ar.decode()
print(s)
print(type(s))
```

Function Defintion In Python

```python
def is_even(num : int) -> bool:
    return num % 2 == 0
```

Function Application

```python
is_even(2)
```

```python
is_even(3)
```

```python
def even_or_odd(num : int) -> str:
    if num % 2 == 0:
        return "even"
    else:
        return "odd"
```

```python
even_or_odd(2)
```

```python
even_or_odd(3)
```
