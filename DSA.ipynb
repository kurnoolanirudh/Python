```python
class SllNode:
    def __init__(self, data, next_node = None):
        self.data = data
        self.next = next_node
```

```python
class DllNode:
    def __init__(self, data, prev_node = None, next_node = None):
        self.data = data
        self.prev = prev_node
        self.next = next_node
```

```python
class BtNode:
    def __init__(self, data, lchild=None, rchild=None):
        self.data = data 
        self.lchild = lchild
        self.rchild = rchild
```
