Contains Duplicate = Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
def contains_duplicate(arr):
    st = set()
    for i in arr:
        if i in st:
            return True 
        st.add(i)
    return False 


nums = [1,2,3,1]
nums1 = [1,2,3,4]
print(contains_duplicate(nums))
print(contains_duplicate(nums1))
```

Valid Anagram = Given two strings s and t, return true if t is an anagram of s, and false otherwise.

```python
def valid_anagram(s, t):
    if len(s) != len(t):
        return False 
    S, T = dict(), dict()
    for i in range(len(s)):
        S[t[i]] = 1 + S.get(s[i], 0)
        T[t[i]] = 1 + T.get(t[i], 0)

    return S == T


s1, s2 = "anagram", "rat"
t1, t2 = "nagaram", "car"
print(valid_anagram(s1, t1))
print(valid_anagram(s2, t2))
```

Two Sum = Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target. You may assume that each input would have exactly one solution, and you may not use the same element twice. You can return the answer in any order.

```python
def two_sum(nums, target):
    num_to_index = dict()
    for i in range(len(nums)):
        diff = target - nums[i]
        if diff in num_to_index:
            return [i, num_to_index[diff]]
        num_to_index[nums[i]] = i

nums = [2,7,11,15]
target = 9

nums1 = [3,2,4]
target1 = 6

nums2 = [3,3]
target2 = 6

print(two_sum(nums, target))
print(two_sum(nums1, target1))
print(two_sum(nums2, target2))
```
