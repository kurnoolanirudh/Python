```python
import pandas as pd
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
ages = data['Age']
all_devs = data['All_Devs']
python = data["Python"]
js = data["JavaScript"]
```

To fill the area under the curve with a color

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
plt.fill_between(ages, python, alpha=0.25) # aplha indicates how transparent the fill should be
# alpha = 0 (completely transparent) ; alpha = 1 (not at all transparent)
plt.plot()
plt.legend()
plt.show()
```

to know where the filled plot crosses a median

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")

mdn = 57287

plt.fill_between(ages, python, mdn, alpha=0.25)
plt.plot()
plt.legend()
plt.show()
```

To apply a condition to the fill

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, mdn, where=(python > mdn), interpolate=True, alpha=0.25) # aplha indicates how transparent the fill should be
# alpha = 0 (completely transparent) ; alpha = 1 (not at all transparent)
plt.plot()
plt.legend()
plt.show()
```

Apply colors conditionally

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, mdn, where = (python > mdn), color="green", interpolate=True, alpha=0.25)
plt.fill_between(ages, python, mdn, where = (python < mdn), color="red", interpolate=True, alpha=0.25)
plt.plot()
plt.legend()
plt.show()
```

to fill the area between the plot

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, all_devs, where = (python > all_devs), color="green", interpolate=True, alpha=0.25, label="Above Average")
plt.fill_between(ages, python, all_devs, where = (python <= all_devs), color="red", interpolate=True, alpha=0.25, label="Below Average")
plt.plot()
plt.legend()
plt.show()
```

Histograms

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages, bins=5, edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("Number")
plt.title("Number Of People By Age")
plt.plot()
plt.show()
```

Giving a custom list of bins 

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages, bins=[10, 20, 30, 40, 50, 60, 70], edgecolor="black")
plt.title("Number Of People By Age")
plt.xlabel("ages")
plt.ylabel("count")
plt.plot()
plt.show()
```

Data can be excluded by restricting bins

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.hist(ages, bins=[20, 30, 40, 50], edgecolor="black")
plt.title("Number Of People By Age")
plt.xlabel("ages")
plt.ylabel("count")
plt.plot()
plt.show()
```

```python
plt.hist(data1["Age"], bins=[i for i in range(10, 100, 10)], edgecolor="black", log=True)

median_age = 29
plt.axvline(median_age, label="Median Age", color="red" ,linewidth=2)

plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Vs Correspondents")
plt.plot()
plt.show()
```

Scatter Plot

```python
import random
```

```python
x = [random.randrange(0, 10) for _ in range(20)]
y = [random.randrange(0, 10) for _ in range(20)]
colors = [random.randrange(0, 10) for _ in range(20)]
sizes = [30 * random.randrange(0, 10) for _ in range(20)]
plt.scatter(x, y, s=sizes, c=colors, cmap="Greens", edgecolors="black", linewidths=2)

cbar = plt.colorbar()
cbar.set_label("intensity")
plt.grid()
plt.show()
```

log scale scatter plot

```python
x = [random.randrange(0, 1000) for _ in range(1000)]
y = [random.randrange(0, 1000) for _ in range(1000)]
colors = [random.randrange(0, 1000) for _ in range(1000)]
sizes = [0.1 * random.randrange(0, 1000) for _ in range(1000)]
plt.scatter(x, y, s=sizes, c=colors, cmap="Greens", edgecolors="black", linewidths=2)
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label("intensity")
plt.grid()
plt.show()
```

Time Series Data

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot(dates, y)
```

```python
plt.plot(dates, y)
plt.gcf().autofmt_xdate()
```

To use custom date formats

```python
plt.plot(dates, y)
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```
