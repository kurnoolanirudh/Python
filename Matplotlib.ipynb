```python
from matplotlib import pyplot as plt
```

```python
x = [i for i in range(10)]
y = [2 * i  for i in x]

plt.plot(x, y)
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = [i for i in range(5)]
y = [2 * i  for i in x]
z = [i ** 2 for i in x]

plt.plot(x, y)
plt.plot(x, z)
plt.title("x vs y and x vs z")
plt.xlabel("x")
plt.ylabel("y, z")
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
ages = data['Age']
all_devs = data['All_Devs']
python = data["Python"]
js = data["JavaScript"]
```

To fill the area under the curve with a color

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
plt.fill_between(ages, python, alpha=0.25) # aplha indicates how transparent the fill should be
# alpha = 0 (completely transparent) ; alpha = 1 (not at all transparent)
plt.plot()
plt.legend()
plt.show()
```

to know where the filled plot crosses a median

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")

mdn = 57287

plt.fill_between(ages, python, mdn, alpha=0.25)
plt.plot()
plt.legend()
plt.show()
```

To apply a condition to the fill

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, mdn, where=(python > mdn), interpolate=True, alpha=0.25) # aplha indicates how transparent the fill should be
# alpha = 0 (completely transparent) ; alpha = 1 (not at all transparent)
plt.plot()
plt.legend()
plt.show()
```

Apply colors conditionally

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, mdn, where = (python > mdn), color="green", interpolate=True, alpha=0.25)
plt.fill_between(ages, python, mdn, where = (python < mdn), color="red", interpolate=True, alpha=0.25)
plt.plot()
plt.legend()
plt.show()
```

to fill the area between the plot

```python
plt.plot(ages, all_devs, linestyle="--", color="#444444", label="All Devs")
plt.plot(ages, python, label="Python")
plt.xlabel("Age")
plt.ylabel("Salaries")
plt.title("Age vs Salary")
mdn = 57287
plt.fill_between(ages, python, all_devs, where = (python > all_devs), color="green", interpolate=True, alpha=0.25, label="Above Average")
plt.fill_between(ages, python, all_devs, where = (python <= all_devs), color="red", interpolate=True, alpha=0.25, label="Below Average")
plt.plot()
plt.legend()
plt.show()
```
