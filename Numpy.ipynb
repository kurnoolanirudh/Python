Linear Algebra

to get eigen values and eigen vectors

```python
mt = np.array([[1, 2], [3, 4]])
```

```python
eigen_values, eigen_vectors = np.linalg.eig(mt)
print(eigen_values)
print(eigen_vectors) # column vectors
```

eigen_value * eigen_vector = matrix * eigen vector

```python
print(eigen_vectors[:,0] * eigen_values[0])
print(mt @ eigen_vectors[:, 0])
```

```python
i1 = eigen_vectors[:,0] * eigen_values[0]
i2 = mt @ eigen_vectors[:, 0]
```

to compare two floating point values

```python
np.allclose(i1, i2)
```
