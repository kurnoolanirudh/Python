to set the index to be certain column when reading the csv

```python
data = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_public.csv", index_col="ResponseId")
```

to apply a filter 

```python
data[data["Employment"] == "Retired"]
```

```python
data.loc[data["Employment"] == "Retired"]
```

get only when the WorkExp column after the filter is applied

```python
data.loc[data["Employment"] == "Retired", "WorkExp"]
```

To combine multiple filters in pandas ( & = and ; | = or ; ~ = not )

```python
data[(data["Employment"] == "Retired") & (data["WorkExp"] <= 10)]
```

```python
data[(data["Employment"] == "Retired") & (data["WorkExp"] <= 5)]
```

```python
data[(data["Employment"] == "Retired") | (data["Employment"] == "Unemployed")]
```

```python
data[~(data["Employment"] == "Retired")]
```

```python
data[~(data["Employment"] == "Retired")].head()
```

```python
high_comp = data["ConvertedCompYearly"] > 2_000_000
```

```python
data[high_comp]
```

To select rows where some column's value is in a list

```python
countries = ["India", "Canada"]
india_and_canada = data["Country"].isin(countries)
data[india_and_canada]
```

```python
python_devs = data["LanguageHaveWorkedWith"].str.contains("Python", na=False)
data[python_devs]
```
